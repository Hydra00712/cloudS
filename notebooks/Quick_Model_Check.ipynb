# Quick Model Check - Engagement Prediction

Fast verification of model registry and experiments

```python
from azure.ai.ml import MLClient
from azure.identity import DefaultAzureCredential

# Connect
ml_client = MLClient.from_config(credential=DefaultAzureCredential())
print(f"✅ Workspace: {ml_client.workspace_name}")
```

```python
# Check model
model = ml_client.models.get("engagement-gb-model", label="latest")
print(f"✅ Model: {model.name} v{model.version}")
print(f"   MAE: {model.tags.get('mae')}")
print(f"   RMSE: {model.tags.get('rmse')}")
print(f"   R²: {model.tags.get('r2')}")
```

```python
# Check experiments
jobs = list(ml_client.jobs.list())[:3]
print(f"✅ Found {len(jobs)} recent jobs")
for job in jobs:
    print(f"   - {job.name}: {job.status}")
```

```python
# Check compute
compute = ml_client.compute.get("cpu-cluster-small")
print(f"✅ Compute: {compute.name}")
print(f"   Type: {compute.type}")
print(f"   State: {compute.provisioning_state}")
```
